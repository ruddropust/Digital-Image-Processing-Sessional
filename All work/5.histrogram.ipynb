```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread("cat1.jpg") 
img = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img2 = np.copy(img)
row = img2.shape[0]
column = img2.shape[1]
```

```python
img2=img.flatten()
img2
```

```python
img2=np.sort(img2)
img2.shape
```

```python
hist = np.zeros(256)
for i in img2:
    hist[i] = hist[i]+1
```

```python
hist = hist/(row*column)
```

```python
hist_difference = np.array([x if i%5 == 0 else 0 for i,x in enumerate(hist)])
plt.bar(np.arange(256),hist_difference)
```
