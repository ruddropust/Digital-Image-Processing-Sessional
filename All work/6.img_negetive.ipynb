```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image = cv.imread("plant.jpg")
image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
plt.axis('off')
print(image.shape)
```

```python
def image_negative(img):
    negative = 255-img
    return negative
```

```python
img_neg = image_negative(img)
img_neg
plt.imshow(img_neg, cmap="gray")
plt.axis("off")
```
